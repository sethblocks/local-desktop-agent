# Testing for improved search ability

```python
import cloudscraper
from bs4 import BeautifulSoup
url = "https://www.reddit.com/r/gnome/comments/1k7uoq7/set_gdm_accent_colors_to_match_gnome_shell/"
scraper = cloudscraper.create_scraper()
content = scraper.get(url).text
soup = BeautifulSoup(content)
text = soup.text
while text.find("\n\n") != -1:
    text = text.replace("\n\n", "\n")
print(text)
```
